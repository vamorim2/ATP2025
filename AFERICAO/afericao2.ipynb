```python
# a

def listar_unidades(hospital):
    unidades = []
    for elem in hospital:
        for elem1 in elem[0]:
            unidades.append(elem1[0])
    return unidades

print(listar_unidades(hospital))
```

```python
# b

def camas_disponiveis(hospital):
    camasdisponiveis = []
    for elem in hospital:
        for elem1 in elem[2]:
            if elem1[1] == True:
                camasdisponiveis.append(elem1[0])
    return camasdisponiveis

print(camas_disponiveis(hospital))
```
